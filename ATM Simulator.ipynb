```python
class Customer:
    def __init__(self,first,last,pin,balance):
        self.first = first
        self.last = last
        self.pin = pin
        self.balance = balance
        
customer_1 = Customer("Mayur","Gaikwad",7243,25000)
customer_2 = Customer("Swapnil","Kamble",2685,37000)
customer_3 = Customer("Ganesh","Jadhav",5194,42000)
customer_4 = Customer("Viraj","Vadapkar",9617,30000)

```

```python
print("...Welcome to HDFC ATM...")
print("...Please Enter the pin...")

pin = int(input("Enter the Pin: "))

if pin==(customer_1.pin or customer_2.pin or customer_3.pin or customer_4.pin):
    print("Account authorised")
else:
    print("Pin is incorrect")

functionality = input(f"Enter the number to proceed [Check balance(0)/Withdrawal(1)/Deposit(2)/Exit(3):] ")

if pin==customer_1.pin:
    if functionality == "0":
        print("The current balance is: Rs.",customer_1.balance)
    elif functionality == "1":
        withdrawal=int(input("Enter the amount: Rs."))
        if withdrawal>customer_1.balance:
            print("Not enough balance")
        else :
            print(withdrawal,"Rs. withdrawn successfully")
            print(f"Updated balance is: Rs.{customer_1.balance-withdrawal}")
    elif functionality == "2":
        deposit=int(input("Enter the amount: Rs. "))
        print(deposit,"Deposited successfully")
        print(f"Updated balance is: Rs.{customer_1.balance+deposit}")
    elif functionality =="3":
        print("Exiting...")
        print("You have been successfully logged off")
    else :
        print("Select the no.0/1/2/3 for further activity ")

elif pin==customer_2.pin:
    if functionality == "0":
        print("The current balance is: Rs.",customer_2.balance)
    elif functionality == "1":
        withdrawal=int(input("Enter the amount:Rs. "))
        if withdrawal>customer_1.balance:
            print("Not enough balance")
        else :
            print(withdrawal,"Rs. withdrawn successfully")
            print(f"Updated balance is: Rs.{customer_2.balance-withdrawal}")
    elif functionality == "2":
        deposit=int(input("Enter the amount: Rs. "))
        print(deposit,"Deposited successfully")
        print(f"Updated balance is: Rs.{customer_2.balance+deposit}")
    elif functionality =="3":
        print("Exiting...")
        print("You have been successfully logged off")
    else :
        print("Select the no.0/1/2/3 for further activity ")
        
elif pin==customer_3.pin:
    if functionality == "0":
        print("The current balance is: Rs.",customer_3.balance)
    elif functionality == "1":
        withdrawal=int(input("Enter the amount: Rs. "))
        if withdrawal>customer_1.balance:
            print("Not enough balance")
        else :
            print(withdrawal,"Rs. withdrawn successfully")
            print(f"Updated balance is: Rs.{customer_3.balance-withdrawal}")
    elif functionality == "2":
        deposit=int(input("Enter the amount: Rs. "))
        print(deposit,"Deposited successfully")
        print(f"Updated balance is: Rs.{customer_3.balance+deposit}")
    elif functionality =="3":
        print("Exiting...")
        print("You have been successfully logged off")
    else :
        print("Select the no.0/1/2/3 for further activity ")
        
elif pin==customer_4.pin:
    if functionality == "0":
        print("The current balance is: Rs.",customer_4.balance)
    elif functionality == "1":
        withdrawal=int(input("Enter the amount: Rs. "))
        if withdrawal>customer_1.balance:
            print("Not enough balance")
        else :
            print(withdrawal,"Rs.withdrawn successfully")
            print(f"Updated balance is: Rs.{customer_4.balance-withdrawal}")
    elif functionality == "2":
        deposit=int(input("Enter the amount:Rs. "))
        print(deposit,"Deposited successfully")
        print(f"Updated balance is: Rs.{customer_4.balance+deposit}")
    elif functionality =="3":
        print("Exiting...")
        print("You have been successfully logged off")
    else :
        print("Select the no.0/1/2/3 for further activity ")
        
else:
    print("Enter the correct pin.")



```
